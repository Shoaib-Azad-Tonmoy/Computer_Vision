```python
def bubblesort(numbers):
    for n in range(len(numbers) - 1, 0, -1):
        print(f"iteration {len(numbers) - n}:")
        print(numbers)
        for i in range(n):
            if numbers[i] > numbers[i + 1]:
                numbers[i], numbers[i + 1] = numbers[i + 1], numbers[i]
            print(f'step {i + 1}:')
            print(numbers)

        print("******************")
        print()

numbers = [34, 56, 23, 43, 57, 81, 28, 93, 51, 77]
print("Unsorted list is,")
print(numbers)
print()
bubblesort(numbers)
print("Sorted Array is, ")
print(numbers)
```
